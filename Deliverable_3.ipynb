```python
#Step (1): Import the pandas, numpy, and seaborn libraries
import pandas as pd

import numpy as np

import seaborn as sns
```

```python
#Step (2): Use the code line below to load the "iris" dataset and store it as a data frame
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Step (3): Drop the "species" column in the iris data frame
iris.drop(labels = 'species', axis = 1, inplace = True)
iris.head()
```

```python
#Step (4): Add two more columns to the iris data frame
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
#Step (5): Create a new data frame with the summary stats for the iris dataset
mean = [np.mean(iris['sepal_length']),
        np.mean(iris['sepal_width']),
        np.mean(iris['sepal_sum']),
        np.mean(iris['petal_length']),
        np.mean(iris['petal_width']),
        np.mean(iris['petal_sum'])]

summary_dict = {'mean': mean}

summary_df = pd.DataFrame(data = summary_dict)

summary_df
```

```python
# Describes entire dataset
iris.describe()
```

```python
#Final Product
summary_stats = pd.DataFrame({
    'mean': iris.mean(),
    'std': iris.std(),
    'min': iris.min(),
    'max': iris.max()
})

# Reorder the columns
summary_stats = summary_stats.reindex(['sepal_length', 'sepal_width', 'sepal_sum', 'petal_length', 'petal_width', 'petal_sum'])

print(summary_stats)
```
